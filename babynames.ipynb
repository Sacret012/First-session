```python

years = range(1880, 1883)
pieces = []
for i in years:
    df1 = pd.read_table('D:\yob%d.txt'%i, sep = ',', engine = 'python' , names=['name', 'sex', 'number'])
    df2 = pd.read_table('D:\yob%d.txt'%i, sep = ',', engine = 'python' , names=['name', 'sex', 'number'])   
    df=pd.concat([df1, df2])
    df['year'] = i
    pieces.append(df)
    data = pd.concat(pieces, ignore_index = True)
data
```

```python

YearsSexCount = data.groupby(['year', 'sex']).size()
YearsSexCount
```

```python

abscissa = []
ordinada_F = []
ordinada_M = []

for i in range(1880, 2011):
    abscissa.append(i)

for i in range(0, 262):
    if i % 2 == 0:
        ordinada_F.append(YearsSexCount.iloc[i])
    else:
        ordinada_M.append(YearsSexCount.iloc[i])
        
plt.plot(abscissa, ordinada_F, label='Famale')
plt.plot(abscissa, ordinada_M, label='Male')
plt.legend()
```
